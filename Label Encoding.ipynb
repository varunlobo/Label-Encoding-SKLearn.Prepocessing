## Turn categorical column into numerical using LabelEncoder & One-Hot Encoder from Sklearn.Preprocessing library

```python
# import essential libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from csv file

df=pd.read_csv("employee_data.csv")
df.head()
```

```python
#check which columns are numeric (int64) while which are non numeric (object)

df.info()
```

```python
# List column names

df.columns
```

```python
# create a new dataframe with only non numeric column data

df_object=df[['Attrition','BusinessTravel','Department','EducationField',
              'Gender','JobRole','MaritalStatus','Over18','OverTime']]
```

```python
# to confirm all datatypes in this new dataframe is object

df_object.info()
```

```python
# Import LabelEncoder from sklearn.preprocessing and apply fit_transform method to convert all object into numberic datatypes

from sklearn.preprocessing import LabelEncoder
df_object.apply(LabelEncoder().fit_transform)
```

## In this notebook we have successfully converted all non-numeric features into numeric so we can use the information in building a more accurate model.
